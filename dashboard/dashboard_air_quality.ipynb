```python
from pathlib import Path
import pandas as pd

data_path = Path("data/analytics/air_quality_analytics.csv")

if data_path.exists():
    print("✅ Cargando datos desde analytics:", data_path)
    df = pd.read_csv(data_path)
else:
    df = pd.DataFrame({
        "parameter": ["pm25", "pm10", "no2", "o3", "co"],
        "avg_value": [23.5, 35.2, 18.1, 12.4, 0.9],
        "city": ["Ciudad A", "Ciudad B", "Ciudad C", "Ciudad D", "Ciudad E"]
    })

df.head()
```

```python
overall_mean = df["avg_value"].mean()
print(f"Overall mean pollution level: {overall_mean:.2f}")
```

```python
fig1 = px.bar(
    df,
    x="parameter",
    y="avg_value",
    title="Average Pollution Levels by Parameter",
    labels={"parameter": "Pollutant", "avg_value": "Average Value"},
    color="parameter",
)
fig1.show()
```

```python
df_sorted = df.sort_values("avg_value", ascending=False)

fig2 = px.bar(
    df_sorted,
    x="avg_value",
    y="city" if "city" in df_sorted.columns else "parameter",
    orientation="h",
    title="Pollution Ranking",
    labels={"avg_value": "Average Value"},
)
fig2.show()
```
