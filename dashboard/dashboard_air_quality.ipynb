```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Asegurar modo inline para que las figuras se incrusten en la notebook
%matplotlib inline

data_path = Path("data/analytics/air_quality_analytics.csv")

if data_path.exists():
    print("âœ… Cargando datos desde analytics:", data_path)
    df = pd.read_csv(data_path)
else:
    df = pd.DataFrame({
        "parameter": ["pm25", "pm10", "no2", "o3", "co"],
        "avg_value": [23.5, 35.2, 18.1, 12.4, 0.9],
        "city": ["Ciudad A", "Ciudad B", "Ciudad C", "Ciudad D", "Ciudad E"]
    })

df.head()
```

```python
overall_mean = df["avg_value"].mean()
print(f"Overall mean pollution level: {overall_mean:.2f}")
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["parameter"], df["avg_value"])
plt.title("Average Pollution Levels by Parameter")
plt.xlabel("Parameter")
plt.ylabel("Average Value")
plt.tight_layout()
plt.show()
```

```python
df_sorted = df.sort_values("avg_value", ascending=False)

plt.figure(figsize=(6,4))
plt.barh(df_sorted["city"], df_sorted["avg_value"])
plt.title("Pollution Ranking by City")
plt.xlabel("Average Value")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```
