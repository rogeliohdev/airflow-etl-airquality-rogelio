```python
# Ruta espera que el archivo esté en la carpeta data/ del proyecto
data_path = Path("data/air_quality_clean.csv")

# Si estás en Colab, sube el archivo manualmente a la izquierda
# o usa files.upload(). Para el repo, lo importante es el código.
df = pd.read_csv(data_path)

df.head()
```

```python
overall_mean = df["avg_value"].mean()
print(f"Overall mean pollution level: {overall_mean:.2f}")
```

```python
fig1 = px.bar(
    df,
    x="parameter",
    y="avg_value",
    title="Average Pollution Levels by Parameter",
    labels={"parameter": "Pollutant", "avg_value": "Average Value"},
    color="parameter"
)
fig1.show()
```

```python
df_sorted = df.sort_values("avg_value", ascending=False)

fig2 = px.bar(
    df_sorted,
    x="avg_value",
    y="city" if "city" in df_sorted.columns else "parameter",
    orientation="h",
    title="Pollution Ranking",
    labels={"avg_value": "Average Value"},
)
fig2.show()
```
